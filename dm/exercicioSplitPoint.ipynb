```python
import seaborn as sns
df_iris = sns.load_dataset("iris")
```

```python
df_iris15 = df_iris.sample(15, random_state = 1)
df_iris15
```

```python
df_iris15t = df_iris15[['petal_width', 'species']]
df_iris15t
```

```python
df_iris15t.sort_values(by='petal_width')
```

Sp1 recebe o mínimo e sp2  recebe o máximo do intervalo entre as classes e divide por 2

```python
sp1 = 0.75
sp2 = 1.80
df_iris15t.loc[:,('petal_width <= sp1')] = df_iris15t['petal_width'].apply(lambda x : x <= sp1).values
df_iris15t.loc[:,('petal_width <= sp2')] = df_iris15t['petal_width'].apply(lambda x : x <= sp2).values
df_iris15t
```

```python
df_iris15t.sort_values(by='petal_width')
```

```python
helper.ganho_de_informacao(df_iris15t, 'petal_width <= sp1', 'species')
```

```python
helper.ganho_de_informacao(df_iris15t, 'petal_width <= sp2', 'species')
```

Observamos que o maior ganho de informação foi do split point xsp1

.

Por isso, esse é o valor do atributo de petal_width que dividiu o conjunto em duas partes, maximizando o ganho de informação.
