Definição de uma função para normalização linear

```python
def normalizacao_linear(x):
    return (x - np.min(x)) / (np.max(x) - np.min(x));
```

Aplicação da função em uma das colunas do conjunto de dados

```python
new_data = normalizacao_linear(data['sepal length (cm)'])
new_data
```

Aplicação da função em todas as colunas no conjunto de dados

```python
new_data = data.apply(normalizacao_linear, axis = 0);
new_data
```

Ii=[(min(x)+L×i),(min(x)+L×(i+1)−1],para1≤i<k

```python
np.random.seed(1)
idade = np.random.randint(low = 1, high = 101, size = 500)
print(idade)
```

Intervalos de Igual Frequência
Neste tipo de discretização a ideia consiste em manter o mesmo número de valores (ou exemplos) em cada intervalo.

Considerando k
o número de intervalos e n

o número de exemplos:

    1.Calculamos N, o número de exemplos por intervalo, como N=nk

    2.Calculamos o ranking dos valores da variável x

    3.Aplicamos o método de discretização de igual tamanho sobre o ranking dos elementos

```python
K = 4
N = np.round(len(idade) / K)

from scipy.stats import rankdata
x = rankdata(idade)

L = (np.max(x) - np.min(x) + 1) / K
intervalos = (x - np.min(x)) // L

intervalos_igual_frequencia = intervalos.astype(int)
print(intervalos_igual_frequencia)
```

```python
from collections import Counter
Counter(intervalos)
```

```python
df = pd.DataFrame({"idade": idade, "rank": x, "intervalos": intervalos_igual_frequencia})
df
```

Conjunto de Dados Original

```python
data = pd.DataFrame(iris['data'], columns=iris['feature_names'], copy=True)
data
```

Índices para remover (to_remove) escolhidos aleatoriamente:

```python
import random
random.seed(1)
to_remove = random.sample(range(0, 150), 30)
print(to_remove)
```

Armazenamos em uma variável os valores antes de serem removidos

```python
missing_true = data.loc[to_remove, 'petal length (cm)'].values
print(missing_true)
data.loc[to_remove, 'petal length (cm)'] = None
```

Conjunto de dados, com valores removidos

```python
data
```

Vamos separar as instâncias que conhecemos os valores de todos os atributos (train) e as instâncias que possuem os valores ausentes (missing)

```python
train = data.drop(to_remove, axis=0)
missing = data.loc[to_remove,:]
missing.head()
```

A partir do conjunto train vamos construir uma regressão para predizer a variável resposta ou target 'sepal length (cm)' utilizando os outros três atributos.

Ajustar uma função na forma y=f(X),
y é a variável resposta ou target. y_train representa y.
X são outros três atributos que podem nos ajudar a prever y. X_train representa X.

```python
target = 'petal length (cm)'
y_train = train[target]
X_train = train.drop(target, axis=1)
```

Instanciamos um regressor linear e ajustamos a regressão aos dados.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print('Coeficientes: ', model.coef_)
print('Intercepto: ', model.intercept_)
```

```python
def regression_to_equation (columns, coefs, intercept, nround = 3):
    coefs = np.round(coefs, nround).astype(str)    
    intercept = np.round(intercept, nround).astype(str)    
    str1 = " + ".join( [(coef + ' x '+ column) for column, coef in zip(columns, coefs)] )        
    return 'y = ' + str1 + ' + ' + intercept;
```

```python
regression_to_equation(X_train.columns, model.coef_, model.intercept_)
```

Agora vamos tentar adivinhar os valores ausentes

```python
X_missing = missing.drop(target, axis=1)

y_missing_pred = model.predict(X_missing)
y_missing_pred
```

A função abaixo permite desenhar um scatter plot evidenciando o quadrado das diferenças.

```python
def scatter_square_diff ( dt, ax ) :
    
    from matplotlib.patches import Rectangle
    dt['diffX'] = dt.X - dt.X.mean()
    dt['diffY'] = dt.Y - dt.Y.mean()
   
    ax.scatter(dt.X, dt.Y)
    ax.hlines(dt.Y.mean(), xmin = dt.X.min(), xmax = dt.X.max(), linestyles="--", label='Y.mean')
    ax.vlines(dt.X.mean(), ymin = dt.Y.min(), ymax = dt.Y.max(), linestyles="--", label='X.mean')

    for i in range(0,len(dt)):
        color  = 'blue' if (dt.diffX.iloc[i] * dt.diffY.iloc[i] > 0 ) else 'red'    
        rect = Rectangle( 
            xy = (dt.X.iloc[i], dt.Y.iloc[i]), 
            width = -dt.diffX.iloc[i], 
            height = -dt.diffY.iloc[i], 
            linewidth = 1, edgecolor = color,facecolor = color, alpha = 0.2)
        ax.add_patch(rect)
   
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.legend()    
    return ax
```

```python
fig, ax = plt.subplots(1, 2, figsize=(17, 6) )

ax[0] = scatter_square_diff(dt, ax[0])
ax[0].set_title('dt')

ax[1] = scatter_square_diff(dt2, ax[1])
ax[1].set_title('dt2')

plt.show()
```



    Os quadrados azuis representam as multiplicações do somatório que retornam valores positivos.
    Os quadrados vermelhos representam as multiplicações do somatório que retornam valores negativos.



No conjunto de dados dt observamos que a soma desses quadrados é:

```python
dt_soma_quadrados = np.sum((dt.X - dt.X.mean()) * (dt.Y - dt.Y.mean()))
print('dt_soma_quadrados: ', dt_soma_quadrados.round(2))
dt_cov = dt_soma_quadrados / (n - 1)
print('dt_cov: ', dt_cov.round(2))
```

á no conjunto de dados dt2 observamos que a soma desses quadrados é:

```python
dt2_soma_quadrados = np.sum((dt.X - dt.X.mean()) * (dt2.Y - dt2.Y.mean()))
print('dt_soma_quadrados: ', dt2_soma_quadrados.round(2))
dt2_cov = dt2_soma_quadrados / (n - 1)
print('dt2_cov: ', dt2_cov.round(2))
```

A covariância entre as variáveis X e Y do conjunto de dados dt é positiva e algumas vezes maior do que a covariância de X e Y

em dt2 que, por sinal, foi negativa.

Valores de covariância positivos indicam que valores acima da média de uma variável estão associados com valores médios acima da outra variável.

Valores de covariância negativos indicam que valores acima da média de uma variável estão associados com valores médios abaixo da outra variável.

Valores muito diferentes de zero indicam que valores associação entre as variáveis, já valores muito próximos de zero indica pouca associaçao entre as variáveis.

No coeficiente de determinação estamos apenas interessados em saber a dependência entre as variáveis. Seja ρ o coeficiente de correlação entre as variáveis X e Y, o coeficiente de determinação, também chamado de R2, é calculado pelo quadrado do coeficente de correlação, ρ2

Dizemos também que o valor da variável X explica em 98% o valor da variável Y.

Exemplo de que os atributos mesmo estando ordenados não altera a entropia

```python
ordenados = sorted(random_class)
ordenados
```

```python
def entropia_valores(valores):
    import collections
    counter = collections.Counter(valores)
    vp = [ (n/len(valores)) for n in counter.values() ]    
    return entropia(vp)
```

```python
def entropia (vp):
    somatorio = 0;
    for p in vp :
        somatorio += 0 if (p == 0) else p * np.log2(p)    
    return somatorio * (-1)
```
